```r
x<-c(6, 11, 16, 24, 41)
y<-c(1800, 2200, 2400, 2400, 2200)
relation<-lm(y~x+I(x^2))
print(relation)
summary(relation)
```

```r
x_seq <- seq(from=6,to=41, by=0.1)
new_relation<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation[i]<-predict(relation, x_new)
}
plot(x,y,col="blue", main="Linear Regression Example",lines(x_seq, new_relation),cex=1.3,pch=16,xlab="x",ylab="y")
```

```r
confint(relation, level=0.95)
```

```r
plot(fitted(relation), residuals(relation))
```
